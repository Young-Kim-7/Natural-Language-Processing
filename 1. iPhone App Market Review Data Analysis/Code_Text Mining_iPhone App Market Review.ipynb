## **1.2. Lexicon-Based Sentiment Analysis (Unsupervised Machine Learning)**

We will use the VADER lexicon available through the NLTK module.

```python
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

Let's score all reviews as well as split them into positive, negative, or neutral.

```python
def analyze_sentiment_vader_lexicon(review, threshold = 0.1, verbose = False):
    scores = analyzer.polarity_scores(review)

    # Classify consumers' sentiments into positive, negative, and neutral.
    if scores['compound'] >= threshold:
        binary_sentiment = 'positive'
    elif scores['compound'] <= -threshold:
        binary_sentiment = 'negative'
    else:
        binary_sentiment = 'neutral'

    if verbose:
        print('VADER Polarity (Binary):', binary_sentiment)
        print('VADER Score:', round(scores['compound'], 2))
    return binary_sentiment,scores['compound']
```

```python
VADER_polarity_test = [analyze_sentiment_vader_lexicon(review, threshold=0.1) for review in NORM_reviews]
VADER_polarity_test_df = pd.DataFrame(VADER_polarity_test, columns = ['VADER Polarity','VADER Score'])
VADER_polarity_test_df.head()
```

## **2.2. Define two functions what will display the results of fitting a topic model.**

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

def get_topic_words(vectorizer, lda_model, n_words):
    keywords = np.array(vectorizer.get_feature_names_out())
    topic_words = []
    for topic_weights in lda_model.components_:
        top_word_locs = (-topic_weights).argsort()[:n_words]
        topic_words.append(keywords.take(top_word_locs).tolist())
    return topic_words
```

## **2.4. Latent Dirichlet Allocation**
We will be using the LatentDirichletAllocation function. The function has the following parameters to be set:

1. Number of topics to model: n_components
2. Parameter vector for the Dirichlet distribution for topics in a document: doc_topic_prior
3. Parameter vector for the Dirichlet distribution for words in a topic: topic_word_prior

Now, let's set the parameters and estimate the topic model:


```python
lda_positive_corpus = LatentDirichletAllocation(n_components=4, max_iter=500,
                                           doc_topic_prior = 0.7,
                                           topic_word_prior = 0.7).fit(bow_positive_reviews)
```

```python
lda_negative_corpus = LatentDirichletAllocation(n_components=4, max_iter=500,
                                           doc_topic_prior = 0.7,
                                           topic_word_prior = 0.7).fit(bow_negative_reviews)
```

```python
lda_neutral_corpus = LatentDirichletAllocation(n_components=4, max_iter=500,
                                           doc_topic_prior = 0.7,
                                           topic_word_prior = 0.7).fit(bow_neutral_reviews)
```

## **2.8. Investigate whether ratings serve as an excellent indicator for sentiment classification compared to reviews**

For comparison with reviews, we assume that five or four ratings are positive, three is neutral, and finally two or one ratings are negative.

```python
# Define a function that categorizes ratings
def categorize_rating(rating):
    if rating >= 4:
        return 'positive'
    elif rating == 3:
        return 'neutral'
    else:
        return 'negative'

# Apply the function to the 'ratings' column to create a new 'Sentiment' column
combined_df['Sentiment in rating'] = combined_df['Ratings'].apply(categorize_rating)
```

We noticed that the accuracy rate of estimation on neutral reviews is so low. There are some assumptions on this result. First, it would be due to sparse data for neutral ratings. From the matrix, it looks like there are significantly fewer neutral ratings compared to positive one. This imbalance can lead to less data for the model to learn what truly constitutes neutrality in the context of your dataset, thereby affecting the accuracy. Second, it would be due to ambiguity in text. Textual data often contains ambiguities and subtleties that are hard to capture with sentiment analysis tools. Phrases that might seem neutral to a human could be interpreted differently by VADER due to the presence of certain keywords.
